```python
import pandas as pd
import requests
from io import StringIO

# URL of the dataset
url = "https://dataverse.harvard.edu/api/access/datafile/:persistentId?persistentId=doi:10.7910/DVN/BZBKJP/UTKTNH"

# Make a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Use StringIO to create a file-like object from the response content
    data = StringIO(response.text)

    # Read the data into a DataFrame, specifying the tab as the delimiter
    df1 = pd.read_csv(data, sep='\t')

    # Display the first few rows of the DataFrame
    print(df1.head())
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
# Read the Excel file into a DataFrame
df2 = pd.read_excel('imf-dm-export-20231209.xls')

# Transpose the DataFrame
df2 = df2.T

# Reset the index and set the first row as column headers
df2.columns = df2.iloc[0]
df2 = df2[1:]

# Reset the index to make the years a column
df2.reset_index(inplace=True)

# Rename the columns
df2.columns = ["Year", "Country"] + list(df2.columns[2:])

# Display the modified DataFrame
df2.head()
```
